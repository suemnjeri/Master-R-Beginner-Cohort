## 4. Compare and contrast coord_cartesian VS xlim() or ylim() when zooming in on a histogram. What happens if you leave binwidth unset? What happens if you try and zoom so only half a bar shows?

Answer:

Xlim() and ylim() - These are shortcuts for supplying the limits argument to the individual scales. By default, any values outside the limits will be replaced with NA, therefore dropping observations outside the range from the plot.

coord_Cartesian()- The Cartesian coordinate system is the most familiar, and common, type of coordinate system. Setting limits on the coordinate system will zoom the plot (like you're looking at it with a magnifying glass), and will not change the underlying data like setting limits on a scale will.

```r
#using xlim - removes observations that are not in the set limits
ggplot(diamonds) + geom_histogram(mapping = aes(x = price), binwidth = 700) + 
  xlim(500,10000)

#using coord_cartesian - maintains the observations and kind of zooms in
ggplot(diamonds) + geom_histogram(mapping = aes(x = price), binwidth = 700) + 
  coord_cartesian(xlim = c(500,10000))
```

# 7.4.1 EXERCISES - Missing Values

## 1. What happens to missing values in a histogram? What happens to missing values in a bar chart? Why is there a difference?

Answer: In geom_bar(), the NA is treated as another variable. The 'x' aesthetic requires a categorical/discrete variable and missing values act like another category.

In a geom_histogram(), the 'x' aesthetic variable needs to be numeric, and stat_bin() groups the observations by ranges into bins. Since the numeric value of the 'NA' observation is unknown, they cannot be placed in a particular bin and are dropped.

Both plots are shown below.

```r
#plotting a histogram with NAs, which are dropped
data.frame(value = c(NA,NA,NA,rnorm(1000,0,1))) %>% ggplot() + geom_histogram(mapping = aes(x = value), bins = 50)
```

```r
#plotting a geom_bar with NAs, which are included as a category
ggplot(data = data.frame(type = c('A','A','A','B','B',NA))) + geom_bar(mapping = aes(x = type))
```
